```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/Tottenham_AllSeason.csv')
```

```python
df.head(50)
```

```python
# format date
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.drop(columns='Div', inplace=True)
```

```python
# create column season

df['Season'] = ''

seasons = ["2009-10", "2010-11", "2011-12", "2012-13", "2013-14", "2014-15", "2015-16", "2016-17", "2017-18", "2018-19", "2019-20", "2020-21"]
season_index = 0

for index, row in df.iterrows():
    if index % 38 == 0 and index != 0:
        season_index += 1
    df.at[index, 'Season'] = seasons[season_index]
```

```python
df.head()
```

```python
def matchDecision(row):
    team = 'Tottenham'
    home_team = row['HomeTeam']
    away_team = row['AwayTeam']
    result = row['FTR']
    
    if home_team == team:
        if result == 'H':
            return 'W'
        elif result == 'A':
            return 'L'
        else:
            return 'D'
    
    if away_team == team:
        if result == 'H':
            return 'L'
        elif result == 'A':
            return 'W'
        else:
            return 'D'

wdd = df.apply(matchDecision, axis=1)
df['WDD'] = wdd
```

```python
def halfTimeDecision(row):
    team = 'Tottenham'
    home_team = row['HomeTeam']
    away_team = row['AwayTeam']
    result = row['HTR']
    
    if home_team == team:
        if result == 'H':
            return 'W'
        elif result == 'A':
            return 'L'
        else:
            return 'D'
    
    if away_team == team:
        if result == 'H':
            return 'L'
        elif result == 'A':
            return 'W'
        else:
            return 'D'

wld = df.apply(halfTimeDecision, axis=1)
df['HTD'] = wld
```
